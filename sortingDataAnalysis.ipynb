```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the csv file and store the data as a pandas dataframe
df = pd.read_csv('sortingData.csv')
```

```python
# Get a list of the unique sort names
sorts = df['Sort Name'].unique()

# Loop over each sort
for sort in sorts:
    
    # Filter the data for the current array size
    data = df[df['Sort Name'] == sort]

    # Set up the figure for each sort
    fig, ax = plt.subplots(figsize=(10, 6))

    # Plot the time taken by each sort for each array size, with the number of threads used as the hue
    sns.barplot(x='Array Size', y='Time Taken', hue='Number of Threads Used', data=data)

    # Set the title, legend, and axis labels
    ax.set_title(f'Time Taken by {sort} for Different Array Sizes')
    ax.legend(title='Number of Threads Used', loc='upper left')
    ax.set_xlabel('Array Size (# of Elements)')
    ax.set_ylabel('Time Taken (s)')
    
    # Display the tick labels with powers of 10 instead of the actual values
    ax.set_xticklabels(['$10^1$', '$10^2$', '$10^3$', '$10^4$', '$10^5$', '$10^6$'])

    # Set the y axis to be logarithmic to better show the differences in time taken 
    # Bubble sort takes too long that the other algorithms are not visible on the graph otherwise
    ax.set_yscale('log')

    # Show the plot
    plt.show()
```
